```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer
categorical_variables = ['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']
for variable in categorical_variables:
    df[variable] = df[variable].astype(str)
    labelbinarizer = LabelBinarizer()
    encoded_data = labelbinarizer.fit_transform(df[variable])
    encoded_df = pd.DataFrame(encoded_data, columns=[f'{variable}_{cls}' for cls in labelbinarizer.classes_])
    df = pd.concat([df, encoded_df], axis=1)
df.head()
```
